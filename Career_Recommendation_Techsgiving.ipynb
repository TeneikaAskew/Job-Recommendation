```python


def driversetup():
    options = webdriver.ChromeOptions()
    #run Selenium in headless mode
    options.add_argument('--headless')
    options.add_argument('--no-sandbox')
    #overcome limited resource problems
    options.add_argument('--disable-dev-shm-usage')
    options.add_argument("lang=en")
    #open Browser in maximized mode
    options.add_argument("start-maximized")
    #disable infobars
    options.add_argument("disable-infobars")
    #disable extension
    options.add_argument("--disable-extensions")
    options.add_argument("--incognito")
    options.add_argument("--disable-blink-features=AutomationControlled")

    driver = webdriver.Chrome(options=options)

    driver.execute_script("Object.defineProperty(navigator, 'webdriver', {get: () => undefined});")

    return driver

def pagesource(url, driver):
    driver = driver
    driver.get(url)
    soup = BeautifulSoup(driver.page_source)
    driver.close()
    return soup
```

```python

# URL of the Google Careers page
url = "https://www.google.com/about/careers/applications/jobs/results?location=United%20States&employment_type=FULL_TIME&company=Google&company=YouTube&has_remote=true&target_level=ADVANCED&target_level=MID"

# Set up the Selenium driver
driver = driversetup()

# Load the page
jobs_html = pagesource(url, driver)
jobs_html
```

```python
# Find all divs with the class ObfsIf-eEDwDf and store in an array
jobs = jobs_html.find_all('div', class_='Ln1EL')
jobs
```

```python
import pandas as pd
job_data = []
# Extract job details
for job in jobs:
    try:
        # Extract job title
        title = job.find('h3', class_='QJPWVe')
        title = title.get_text(strip=True) if title else "N/A"

        # Extract locations
        location_tags = job.find_all('span', class_='r0wTof')
        locations = "; ".join(tag.get_text(strip=True) for tag in location_tags) if location_tags else "N/A"

        # Extract remote eligibility
        remote_span = job.find('span', string="Remote eligible")
        remote_eligible = "Yes" if remote_span else "No"

        # Extract minimum qualifications
        qualifications_section = job.find('div', class_='Xsxa1e')
        qualifications = ""
        if qualifications_section:
            qualifications_list = qualifications_section.find_all('li')
            qualifications = "; ".join(item.get_text(strip=True) for item in qualifications_list)

        # Extract job URL
        learn_more_link = job.find('a', class_='WpHeLc VfPpkd-mRLv6 VfPpkd-RLmnJb')
        job_url = f"https://www.google.com/about/careers/applications/{learn_more_link['href']}" if learn_more_link else "N/A"

        # Append job details
        job_data.append({
            'Title': title,
            'Locations': locations,
            'Remote Eligible': remote_eligible,
            'Qualifications': qualifications,
            'Job URL': job_url
        })
    except Exception as e:
        print(f"Error parsing job: {e}")
        continue

# Convert job data to DataFrame
df = pd.DataFrame(job_data)

# Extract Job ID from Job URL
df['Job ID'] = df['Job URL'].str.extract(r'/results/(\d+)-')


# Display the DataFrame
if not df.empty:
    print(df.count())
else:
    print("No job data found.")
df
```

# Add Job Descriptions

```python
import re
import html
from time import sleep

def clean_html(text):
    """Remove all HTML tags from the given text."""
    return re.sub(r"<[^>]+>", "", text)

def extract_qualifications_and_salary_from_text(job_description):
    """Extract minimum/preferred qualifications and salary range using regex."""
    # Regex to find Minimum and Preferred qualifications
    min_match = re.search(
        r"Minimum qualifications[:\-]?\s*(.*?)Preferred qualifications[:\-]?",
        job_description_str,
        re.DOTALL | re.IGNORECASE
    )
    pref_match = re.search(
        r"Preferred qualifications[:\-]?\s*(.*?)(Responsibilities|The US base salary|<h3>)",
        job_description_str,
        re.DOTALL | re.IGNORECASE
    )

    # Extract and clean qualifications
    minimum_qualifications = clean_html(min_match.group(1).strip()) if min_match else "N/A"
    preferred_qualifications = clean_html(pref_match.group(1).strip()) if pref_match else "N/A"

    # Regex to extract salary range
    salary_match = re.search(
        r"The US base salary range for this full-time position is \$([\d,]+)-\$([\d,]+)",
        job_description
    )

    if salary_match:
        salary_low = salary_match.group(1).replace(",", "")
        salary_high = salary_match.group(2).replace(",", "")
    else:
        salary_low, salary_high = "N/A", "N/A"

    # Truncate job description at salary section
    job_description_truncated = re.split(r"The US base salary|Learn more about", job_description, 1)[0].strip()

    return minimum_qualifications, preferred_qualifications, salary_low, salary_high, job_description_truncated

def fetch_job_details(row):
    """Fetch job details (description, locations, apply link, qualifications, salary) from job page."""
    job_url = row['Job URL']
    if job_url == "N/A":
        return None, None, None, None, None, None, None  # Skip invalid URLs

    try:
        sleep(3)  # Allow time for the page to load
        # Set up the Selenium driver
        driver = driversetup()
        # Load the page
        job_description_html = pagesource(job_url, driver)

        # Extract job description
        job_description_div = job_description_html.find('div', class_='aG5W3')
        job_description = job_description_div.get_text(strip=True) if job_description_div else "N/A"

        # Convert job description to string for regex
        job_description_str = html.unescape(str(job_description))

        # Extract qualifications and salary
        minimum_qualifications, preferred_qualifications, salary_low, salary_high, job_description_truncated = extract_qualifications_and_salary_from_text(job_description_str)

        # Extract all locations
        locations_div = job_description_html.find('div', class_='xbCN0b KiNied M0TuQ LlMNQd')
        all_locations = locations_div.get_text(strip=True).replace('\n', '; ') if locations_div else "N/A"

        # Extract apply link
        apply_a_tag = job_description_html.find('a', id='apply-action-button')
        apply_link = f"https://careers.google.com{apply_a_tag['href']}" if apply_a_tag and 'href' in apply_a_tag.attrs else "N/A"

        return job_description_truncated, all_locations, minimum_qualifications, preferred_qualifications, salary_low, salary_high, apply_link
    except Exception as e:
        print(f"Error fetching details for URL {job_url}: {e}")
        return "Error", "Error", "Error", "Error", "Error", "Error", "Error"

# Assuming `df` is your DataFrame with a "Job URL" column
jobs_data = []
```

```python
from googleapiclient.discovery import build
from googleapiclient.errors import HttpError
from google.colab import auth
import io

def get_google_doc_content(doc_id):
    """
    Export a Google Doc to plain text and return its content.

    Parameters:
        doc_id (str): The Google Doc file ID (found in its URL).

    Returns:
        str: The plain text content of the document.
    """
    # Authenticate and build the Google Drive API service
    auth.authenticate_user()
    from google.auth import default
    creds, _ = default()
    service = build('drive', 'v3', credentials=creds)

    try:
        # Export the Google Doc to plain text
        request = service.files().export_media(fileId=doc_id, mimeType='text/plain')
        file_content = io.BytesIO(request.execute())
        return file_content.getvalue().decode('utf-8')
    except HttpError as error:
        print(f"An error occurred: {error}")
        return None

doc_id = "1CKGglaXyYad16IFiYSDpGd2ofro5dYmi4eD1JNeHkD4"
doc_content = get_google_doc_content(doc_id)

if doc_content:
    print("Document Content:")
    print(doc_content)
    resume_text = doc_content
else:
    print("Failed to retrieve document content.")
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# Generate Recommendations
def recommend_jobs_from_resume(data, resume_text):
    """
    Generate job recommendations based on the resume text and job descriptions.
    Args:
        data (pd.DataFrame): Job descriptions data.
        resume_text (str): Resume content.
    Returns:
        pd.DataFrame: DataFrame with similarity scores.
    """
    # Prepare job descriptions
    # job_descriptions = all_jobs['Job Description'].fillna("").tolist()

    job_descriptions = all_jobs['Job Description'].fillna("").tolist()
    minimum_qualifications = all_jobs['Minimum Qualifications'].fillna("").tolist()
    preferred_qualifications = all_jobs['Preferred Qualifications'].fillna("").tolist()

    combined_qualifications = []
    for i in range(len(job_descriptions)):
        combined = f"{job_descriptions[i]} \n Minimum Qualifications: {minimum_qualifications[i]} \n Preferred Qualifications: {preferred_qualifications[i]}"
        combined_qualifications.append(combined)

    all_jobs['Full Description'] = combined_qualifications

    # Combine resume with job descriptions
    documents = [resume_text] + combined_qualifications#job_descriptions

    # Vectorize text data
    vectorizer = TfidfVectorizer().fit_transform(documents)
    vectors = vectorizer.toarray()

    # Compute cosine similarity
    cosine_sim = cosine_similarity(vectors)
    similarity_scores = cosine_sim[0][1:]  # Exclude self-comparison

    # Add similarity scores to the DataFrame
    data['Similarity Score'] = similarity_scores

    # Sort jobs by similarity score
    recommended_jobs = data.sort_values(by='Similarity Score', ascending=False)

    return recommended_jobs
```

```python
# Generate Recommendations
def recommend_titles_from_resume(data, resume_text):
    """
    Generate job recommendations based on the resume text and job descriptions.
    Args:
        data (pd.DataFrame): Job descriptions data.
        resume_text (str): Resume content.
    Returns:
        pd.DataFrame: DataFrame with similarity scores.
    """
    # Prepare job descriptions
    job_titles = all_jobs['Title'].fillna("").tolist()

    # Combine resume with job descriptions
    documents = [resume_text] + job_titles

    # Vectorize text data
    vectorizer = TfidfVectorizer().fit_transform(documents)
    vectors = vectorizer.toarray()

    # Compute cosine similarity
    cosine_sim = cosine_similarity(vectors)
    similarity_scores = cosine_sim[0][1:]  # Exclude self-comparison

    # Add similarity scores to the DataFrame
    data['Similarity Score'] = similarity_scores

    # Sort jobs by similarity score
    recommended_titles = data.sort_values(by='Similarity Score', ascending=False)

    # return recommended_titles
    return recommended_titles

# Generate recommendations
recommended_titles = recommend_titles_from_resume(all_jobs, resume_text)

# Display the top recommendations
# print(recommended_titles[['Job ID','Title', 'Similarity Score']])
recommended_titles[['Job ID','Title', 'Similarity Score']].head(15)
```

```python
# Rename similarity columns to avoid confusion
recommended_jobs = recommended_jobs.rename(columns={'Similarity Score': 'Description Similarity'})
recommended_titles = recommended_titles.rename(columns={'Similarity Score': 'Title Similarity'})

# Ensure both DataFrames have the 'Job ID' column
if 'Job ID' not in recommended_titles.columns or 'Job ID' not in recommended_jobs.columns:
    raise ValueError("One of the DataFrames is missing the 'Job ID' column.")

# Merge the DataFrames on 'Job ID'
combined_df = pd.merge(
    recommended_titles[['Job ID', 'Title', 'Title Similarity']],
    recommended_jobs[['Job ID', 'Job Description', 'Description Similarity']],
    on='Job ID',
    how='inner'  # Use 'inner' join to keep only matching Job IDs
)

# Reorganize columns for clarity
combined_df = combined_df[['Job ID', 'Title', 'Title Similarity', 'Job Description', 'Description Similarity']]

# Show the top 5 rows
combined_df.head()
```

```python
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import re

# Load stop words
STOP_WORDS = set(stopwords.words('english')).union({
    'efficient', 'maintain', 'aligned', 'join', 'skilled', 'ensuring',
    'lead', 'using', 'build', 'within', 'objectives', 'performant',
    'tools', 'seeking', 'team', 'highly', 'like'
})

def extract_skills(text, predefined_skills, threshold=0.5, exceptions=None):
    """
    Extract skills from text based on predefined skills and tokenization.
    Handles compound phrases (e.g., joined by 'and') and filters irrelevant large text blocks.
    Args:
        text (str): Input text to extract skills from.
        predefined_skills (list): List of known skills to match against.
        threshold (float): Cosine similarity threshold for fuzzy matching.
        exceptions (list): List of single-character skills or other exceptions to include.
    Returns:
        set: Extracted skills.
    """
    if exceptions is None:
        exceptions = []

    # Tokenize and normalize the text
    text = re.sub(r"[^\w\s]", " ", text.lower())  # Remove punctuation
    tokens = text.split()

    # Filter out stop words and retain exceptions
    tokens = [token for token in tokens if (token not in STOP_WORDS or token in exceptions) and (len(token) > 1 or token in exceptions)]

    # Handle compound phrases joined by "and" or similar
    compound_phrases = re.findall(r"([\w\s]+) and ([\w\s]+)", text)
    split_phrases = set()
    for phrase1, phrase2 in compound_phrases:
        split_phrases.update([phrase1.strip(), phrase2.strip()])

    # Filter out large text blocks that are unlikely to be skills
    MAX_LENGTH = 5  # Maximum number of words in a valid skill
    tokens = [token for token in tokens if len(token.split()) <= MAX_LENGTH]

    # Include exact matches from predefined skills
    matched_skills = set(skill for skill in predefined_skills if skill.lower() in text)

    # Add the compound phrases to matched skills
    matched_skills.update(split_phrases)

    # Approximate matching for remaining tokens
    for skill in predefined_skills:
        for token in tokens:
            if len(skill) > 1 and len(token) > 1:  # Skip empty or one-character inputs
                vectorizer = TfidfVectorizer().fit_transform([skill.lower(), token.lower()])
                similarity = cosine_similarity(vectorizer[0:1], vectorizer[1:2])[0][0] #Find similar skills
                if similarity >= threshold:
                    matched_skills.add(skill)

    # Remove duplicate and irrelevant items
    matched_skills = {skill for skill in matched_skills if len(skill.split()) <= MAX_LENGTH}

    return matched_skills


def identify_skill_gaps(resume_text, job_text, predefined_skills, threshold=0.5, exceptions=None):
    """
    Identify skills in the resume and job description, and compute skill gaps.
    Args:
        resume_text (str): The text of the resume.
        job_text (str): The text of the job description.
        predefined_skills (list): List of known skills to match against.
        threshold (float): Cosine similarity threshold for fuzzy matching.
        exceptions (list): List of single-character skills or other exceptions to include.
    Returns:
        tuple: (skills_in_resume, skills_in_job, missing_skills)
    """
    # Extract skills from the resume and job description
    skills_in_resume = extract_skills(resume_text, predefined_skills, threshold, exceptions)
    skills_in_job = extract_skills(job_text, predefined_skills, threshold, exceptions)

    # Identify missing skills
    missing_skills = skills_in_job - skills_in_resume

    return skills_in_resume, skills_in_job, missing_skills

def fetch_learning_paths(skill):
    """
    Fetch learning paths for the given skill from various learning platforms.
    Args:
        skill (str): The skill for which learning paths are required.
    Returns:
        list: A list of learning paths (course title and link).
    """
    learning_paths = [
        {"title": f"{skill} Basics - Beginner Course (Coursera)", "link": f"https://www.coursera.org/search?query={skill}"},
        {"title": f"{skill} Advanced Topics (edX)", "link": f"https://www.edx.org/search?q={skill}"},
        {"title": f"{skill} Masterclass (Udemy)", "link": f"https://www.udemy.com/courses/search/?q={skill}"},
        {"title": f"{skill} On-Demand Training (Skillsoft)", "link": f"https://www.skillsoft.com/catalog?q={skill}"},
        {"title": f"{skill} Expert-Level Training (Pluralsight)", "link": f"https://www.pluralsight.com/search?q={skill}"},
        {"title": f"{skill} Training by Microsoft", "link": f"https://learn.microsoft.com/en-us/training/browse/?terms={skill}"},
        {"title": f"{skill} Learning Path (AWS)", "link": f"https://aws.amazon.com/training/?search={skill}"},
        {"title": f"{skill} Cloud Certifications (Google Cloud)", "link": f"https://cloud.google.com/training?q={skill}"}
    ]
    return learning_paths

def skill_gap_analysis(resume_text, combined_df, threshold=0.5):
    """
    Perform skill gap analysis and generate upskilling recommendations.
    Args:
        resume_text (str): The text of the resume.
        combined_df (pd.DataFrame): DataFrame containing recommended jobs.
    Returns:
        pd.DataFrame: A DataFrame containing missing skills and learning paths.
    """
    # Extract job descriptions from the combined DataFrame
    job_descriptions = combined_df['Job Description'].fillna("").tolist()

    # Identify skill gaps
    resume_skills, job_skills, missing_skills = identify_skill_gaps(resume_text, job_descriptions, threshold)
    print("Skills in Resume:", resume_skills)
    print("Skills Required by Jobs:", job_skills)
    print("Missing Skills:", missing_skills)

    # Generate upskilling recommendations
    recommendations = []
    for skill in missing_skills:
        learning_paths = fetch_learning_paths(skill)
        for path in learning_paths:
            recommendations.append({"Skill": skill, "Course Title": path["title"], "Course Link": path["link"]})

    return pd.DataFrame(recommendations)


# Example Usage
if __name__ == "__main__":
    # Predefined skill set
    predefined_skills = [
        "Python", "SQL", "Java", "Data Analysis", "Data Modeling",
        "Power BI", "Tableau", "Snowflake", "Redshift", "Azure Synapse",
        "Informatica", "Cloud Platforms", "ETL", "Data Visualization", "R",
        "Data pipelines", "data modeling", "data governance", "Data transformation"
    ]

    # Define single-character exceptions like "R"
    exceptions = ["r"]

    # Example resume and job description
    example_resume = """
    Proficient in Python, SQL, Tableau, and R. Experienced in data analysis and visualization.
    Skilled in Snowflake and Redshift. Familiar with cloud computing concepts.
    """

    example_jd = """
    Looking for a Cloud BI Architect experienced in Snowflake, Redshift, Azure Synapse, and R.
    Must have expertise in Power BI, Tableau, ETL processes, data pipelines and data modeling.
    Proficiency in SQL is required.
    """

    # Perform skill gap analysis
    resume_skills, job_skills, missing_skills = identify_skill_gaps(
        example_resume, example_jd, predefined_skills, threshold=0.5, exceptions=exceptions
    )

    # Display results
    print("Skills in Resume:", resume_skills)
    print("Skills in Job Description:", job_skills)
    print("Missing Skills:", missing_skills)



```

```python
def fetch_learning_paths(skill):
    """
    Fetch learning paths for the given skill from various learning platforms.
    Args:
        skill (str): The skill for which learning paths are required.
    Returns:
        list: A list of learning paths (course title and link).
    """
    learning_paths = [
        {"title": f"{skill} Basics - Beginner Course (Coursera)", "link": f"https://www.coursera.org/search?query={skill}"},
        {"title": f"{skill} Advanced Topics (edX)", "link": f"https://www.edx.org/search?q={skill}"},
        {"title": f"{skill} Masterclass (Udemy)", "link": f"https://www.udemy.com/courses/search/?q={skill}"},
        {"title": f"{skill} On-Demand Training (Skillsoft)", "link": f"https://www.skillsoft.com/catalog?q={skill}"},
        {"title": f"{skill} Expert-Level Training (Pluralsight)", "link": f"https://www.pluralsight.com/search?q={skill}"},
        {"title": f"{skill} Training by Microsoft", "link": f"https://learn.microsoft.com/en-us/training/browse/?terms={skill}"},
        {"title": f"{skill} Learning Path (AWS)", "link": f"https://aws.amazon.com/training/?search={skill}"},
        {"title": f"{skill} Cloud Certifications (Google Cloud)", "link": f"https://cloud.google.com/training?q={skill}"}
    ]
    return learning_paths


def skill_gap_analysis(resume_text, combined_df, predefined_skills, threshold=0.5, exceptions=None):
    """
    Perform skill gap analysis and generate upskilling recommendations.
    Args:
        resume_text (str): The text of the resume.
        combined_df (pd.DataFrame): DataFrame containing recommended jobs.
        predefined_skills (list): List of predefined skills to match against.
        threshold (float): Threshold for skill similarity matching.
        exceptions (list): List of single-character skills or exceptions.
    Returns:
        pd.DataFrame: A DataFrame containing missing skills and learning paths.
    """
    if exceptions is None:
        exceptions = []

    # Extract job descriptions from the combined DataFrame
    job_descriptions = combined_df['Job Description'].fillna("").tolist()

    # Identify skill gaps
    resume_skills, job_skills, missing_skills = identify_skill_gaps(
        resume_text, " ".join(job_descriptions), predefined_skills, threshold, exceptions
    )

    print("Skills in Resume:", resume_skills)
    print("Skills Required by Jobs:", job_skills)
    print("Missing Skills:", missing_skills)

    # Generate upskilling recommendations
    recommendations = []
    for skill in missing_skills:
        learning_paths = fetch_learning_paths(skill)
        for path in learning_paths:
            recommendations.append({"Skill": skill, "Course Title": path["title"], "Course Link": path["link"]})

    return pd.DataFrame(recommendations)


# Example Usage
# Predefined skill set
predefined_skills = [
    "Python", "SQL", "Java", "Data Analysis", "Data Modeling",
    "Power BI", "Tableau", "Snowflake", "Redshift", "Azure Synapse",
    "Informatica", "Cloud Platforms", "ETL", "Data Visualization", "R",
    "Data pipelines", "data modeling", "data governance", "Data transformation"
]

# Define single-character exceptions like "R"
exceptions = ["r"]

# Resume text
resume_text = """
Proficient in Python, SQL, and Tableau. Experienced in data analysis and visualization.
Skilled in project management and Agile methodologies. Familiar with AWS and cloud computing concepts.
"""

# Sample combined DataFrame
combined_df = pd.DataFrame({
    'Job ID': ['12345', '67890'],
    'Title': ['Data Scientist', 'Cloud Engineer'],
    'Title Similarity': [0.9, 0.85],
    'Job Description': [
        "Looking for a candidate skilled in Python, Machine Learning, AWS, and Kubernetes.",
        "The ideal candidate will have expertise in data visualization tools like Power BI and Tableau."
    ],
    'Description Similarity': [0.88, 0.86]
})

# Perform skill gap analysis
recommendations_df = skill_gap_analysis(
    resume_text, combined_df, predefined_skills, threshold=0.5, exceptions=exceptions
)

# Display upskilling recommendations
recommendations_df
```
